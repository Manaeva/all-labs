```python
n = b = int(input('  :'))
c =[]
while b != 0:
    m1 = [int(i) for i in input(' n     :').split()]
    c.append(m1)
    b -= 1

for i in c:
    print(i)
print('')
for j in range(n-1):
    for k in range(j+1,n):
        p = c[k][j]/c[j][j]
        for m in range(j,n-1):
            c[k][m] = int(c[k][m] - c[j][m]*p)
for j in range(n-1,-1,-1):
    for k in range(j,-1,-1):
        p = c[k][j]/c[j][j]
        for m in range(1,n-j):
            c[k][-m] = int(c[k][-m] - c[j][-m]*p)
for i in c:
    print(i)
```
