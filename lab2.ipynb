```python
n = int(input('количество:'))
k = 0
mini = 10**8
m = []
mn = []
while k != n:
    x, y = map(int, input('координаты:').split())
    k += 1 
    m.append([x,y])
for i in range(len(m)):
    g = (m[i][0]**2 + m[i][1]**2)**0.5
    mn.append(g)
    mini = min(mini, g)
for i in range(len(mn)):
    if mn[i]==mini:
        print('Ближайшие к началу координат: №',i+1)
        print('Их расстояние от начала:',mini) 
```

```python
n = 0
res = []
mx = 10**8
while n != -1:
    r = input('Введите результат (для окончания "exit")')
    if r == 'exit':
        break
    res.append(int(r))
for i in range(len(res)):
    mx = min(mx,res[i])
if mx != 10**8:
    print('Лучший результат:',mx, 'единиц(ы) времени') 
else:
    print('Результаты не загружены(')
```

```python
n = 0
m = []
pi = 3.14
mx = 10**8
while n != -1:
    a, b = map(str,input('Значения A и B:').split())
    if a == 'exit' or b == 'exit':
        break
    m.append([int(a),int(b)])
    n+=1
for i in range(len(m)):
    pr = m[i][0]*m[i][1] #площадь прямоугольника
    ring = pi * abs(m[i][0]**2 - m[i][1]**2) #площадь кольца
    p = (m[i][0] + m[i][1] + m[i][1])/2 #полупериметр для формулы Герона
    s = (p*(p-m[i][0])*(p-m[i][1])*(p-m[i][1]))**0.5 #площадь треугольника через формулу Герона
    print(f'A = {m[i][0]} | B = {m[i][1]} | S прямоугольника: {pr} | S кольца: {ring} | S треугольника: {s}') 
```
