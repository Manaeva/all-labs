```python
n = int(input(':'))
k = 0
mini = 10**8
m = []
mn = []
while k != n:
    x, y = map(int, input(':').split())
    k += 1 
    m.append([x,y])
for i in range(len(m)):
    g = (m[i][0]**2 + m[i][1]**2)**0.5
    mn.append(g)
    mini = min(mini, g)
for i in range(len(mn)):
    if mn[i]==mini:
        print('   : ',i+1)
        print('   :',mini) 
```

```python
n = 0
res = []
mx = 10**8
while n != -1:
    r = input('  (  "exit")')
    if r == 'exit':
        break
    res.append(int(r))
for i in range(len(res)):
    mx = min(mx,res[i])
if mx != 10**8:
    print(' :',mx, '() ') 
else:
    print('  (')
```
