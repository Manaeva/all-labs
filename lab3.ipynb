```python
a = [int(i) for i in input('Введите массив через пробел:').split(' ')]
print(f'Исходный{a}')
mn = 10**5
for i in a:
    if i > 0:
        mn = min(i,mn)
for i in a:
    if i == mn:
        a.remove(i)
print(f'Результат:{a}')
```

```python
a = [int(i) for i in input('Введите числа через пробел:').split()]
print(f'Начальный массив:{a}')
for i in range(len(a)-1):
    if a[i]<0:
        amin = a[i]
        for j in range(i+1,len(a)):
            if a[j]<0:
                mn = min(a[j],a[i])
                if mn == a[i]:
                    a[i] = a[j]
                    a[j] = mn
print(f'Результат:{a}')
```

```python
a = [int(i) for i in input('Введите элементы через пробел:').split()]
n = []
print(f'Начальный массив:{a}')
for i in range(len(a)):
    if not(a[i] in n):
        n.append(a[i])
print(f'Результат:{n}')
```
