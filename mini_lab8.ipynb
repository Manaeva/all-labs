```python
string = 'Алиса 567 Манаева 102-+'
summa = 0
print('Найденные числа:')
for i in range(len(string)-1):
    if string[i] in '123456789' and string[i+1] == '0':
        summa += int(string[i]+string[i+1])
        print(string[i]+string[i+1])
    if string[i] in '123456789' and string[i+1] != '0':
        summa += int(string[i])
        print(string[i])
print('Сумма включенных чисел:',summa)
```
