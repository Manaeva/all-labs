```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('19-1.txt') as f:
    z=f.read().splitlines()
    m =[list(map(float, i.split())) for i in z]
x = [i[0] for i in m]
y = [i[1] for i in m]
plt.plot(x,y)
plt.show()
```
