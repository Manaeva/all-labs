```python
import matplotlib.pyplot as plt
from random import *
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

seed(110)
k = 3
n = 20


r1 = 1.5
r2 = 1
r3 = 2
x1, y1 = 5, 4
x2, y2 = 0, 0
x3, y3 = -5, 3

x11 = [uniform(x1+r1, x1-r1) for i in range(n)]
y11 = [uniform(y1+r1, y1-r1) for i in range(n)]

x22 = [uniform(x2+r2, x2-r2) for i in range(n)]
y22 = [uniform(y2+r2, y2-r2) for i in range(n)]

x33 = [uniform(x3+r3, x3-r3) for i in range(n)]
y33 = [uniform(y3+r3, y3-r3) for i in range(n)]


x = x11 + x22 + x33
y = y11 + y22 + y33

coords = list(zip(x, y))
shuffle(coords)

plt.subplot(2,1,1)
plt.scatter([i[0] for i in coords], [i[1] for i in coords])
plt.title('Исходные данные')
plt.subplots_adjust(wspace=0.3, hspace=0.3)

kmeans = KMeans(n_clusters=3, random_state=110)
kmeans.fit(coords)

labels = kmeans.labels_

plt.subplot(2,1,2)
plt.scatter([i[0] for i in coords], [i[1] for i in coords], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', marker='X')
plt.title('Результат работы алгоритма')
plt.show()

silhouette_avg = silhouette_score(coords, labels)
print(f'Средний коэффициент силуэта: {silhouette_avg:.2f}')
```
