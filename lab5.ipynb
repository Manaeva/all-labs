```python
x = []
class matrics():
    def __init__(self,mt):
        self.mt = mt
    def massa(self):
        m = []
        for i in self.mt:
            for j in i:
                m.append(j)
        m1=[]
        mx = 0
        mn = 10**8
        for i in m:
            mx = max(mx,i)
            mn = min(mn,i)
        for i in m:
            if i != mx and i != mn:
                m1.append(i)
        x.append(sum(m1)/len(m1))
ma1 = matrics([[1,2,3],[4,5,6],[7,8,9]])
ma2 = matrics([[10,11,12],[13,14,15],[16,17,18]])
ma3 = matrics([[19,20,21],[22,23,24],[25,26,27]])
ma1.massa()
ma2.massa()
ma3.massa()
print(x)
if x[0] > x[1] and x[1] > x[2]:
    print('Последовательность убывающая')
elif x[0] < x[1] and x[1] < x[2]:
    print('Последовательность возрастающая')
else:
    print('Какая-то последовательность')
```

```python
from math import *

d1 = [] #[0,2]
s1 = 0
s2 = 2
h = 1
while s1 < s2 + h:
    d1.append(s1)
    s1 += h
    
d2 = [] # [-1,1]
x1 = -1
x2 = 1
h = 0.2
while x1 < x2 + h:
    d2.append(x1)
    x1 += h
    
def fsin(x):
    return x**2 - sin(x)

def fe(x):
    return exp(x) - 1
#определитель знаков на промежутках

def sign(x):
    sign = []
    if x == d1:
        for i in x:
            y = fsin(i)
            if fsin(i) > 0:
                sign.append('+')
            else:
                sign.append('-')
    if x == d2:
        for i in x:
            y = fe(i)
            if fe(i) > 0:
                sign.append('+')
            else:
                sign.append('-')
    return sign
    
#счетчик смены знаков

def count(x):
    k = 0
    for i in range(len(sign(x))-1):
        if sign(x)[i] != sign(x)[i+1]:
            k+=1
    return k
print('Количество интервалов смены знаков для первой функции:', count(d1))
print('Количество интервалов смены знаков для первой функции:', count(d2))
```
