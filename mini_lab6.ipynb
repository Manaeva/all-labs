```python
from math import *
import matplotlib.pyplot as plt
import numpy as np

def gradientDescend(func, diffFunc, x0, speed, epochs):
    x_old = x0
    xList = [x0]
    yList = [func(x0)]
    for i in range(epochs):
        x_new = x_old - speed*diffFunc(x_old)
        xList.append(x_new)
        yList.append(func(x_new))
        x_old = x_new
    return xList, yList

f = lambda x: exp(-x**2) + x**4 + 2*x**2
difF = lambda x: -2*x*exp(-x**2) + 4*x**3 + 4*x
x_min = 0
x0 = 3
speed = 0.01
epochs = 100

x1 = gradientDescend(f, difF, x0, speed, epochs)[0]
y1 = gradientDescend(f, difF, x0, speed, epochs)[1]
x2 = np.linspace(0, 3)
y2 = [f(x) for x in x2]

#результат работы алгоритма
plt.scatter(x1, y1)
#точка минимума
plt.plot(x1[-1],y1[-1], 'X', markersize = 15, c='k')
#исходная функция
plt.plot(x2, y2, 'g') 
plt.show()
```
