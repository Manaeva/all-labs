```python
listik = []
good = []

class subject():
    def __init__(self,surname, math, physics, russian):
        self.surname = surname
        self.math = math
        self.physics = physics
        self.russian = russian
        
    def delete(self):
        if int(self.math) != 2 and int(self.physics) != 2 and int(self.russian) != 2:
            good.append([[self.surname],[(int(self.math) + int(self.physics) + int(self.russian))/3]])

n = 0
while n != -1:
    s = [i for i in input('Введите фамилию и 3 оценки через пробел:(чтобы закончить "exit")').split(' ')]
    listik.append(s)
    if s == ['exit']:
      listik.remove(['exit'])
      break

for i in listik:
  for j in range(len(i)-3):
    st = subject(i[j],i[j+1],i[j+2],i[j+3])
    st.delete()
      
for i in range(len(good)-1):
  for j in range(len(good)-i-1):
    if good[j][1] < good[j+1][1]:
      good[j], good[j+1] = good[j+1], good[j]
        
print('Список учащихся, успешно сдавшх экзамены, в порядке убывания среднего балла:')

for i in range(len(good)):
  print(*good[i][0],*good[i][1])
```

```python
class grades():
    def __init__(self, group1, group2, group3):
        self.group1 = group1
        self.group2 = group2
        self.group3 = group3

    def adding(self):
        group = []
        group.append(self.group1)
        group.append(self.group2)
        group.append(self.group3)
        s = 0

        #счет среднего в потоке

        for i in group:
            sumingroup = 0
            for j in i:
                sgr = sum(j) / len(j)
                sumingroup += sgr / len(i)
            s += sumingroup
        print('Среднее в потоке:', s / len(group))

        #счет среднего по группам
        k = 0
        everygroup= []
        for i in group:
            k +=1
            sr = 0
            for j in i:
                sr += sum(j)/len(j)
            everygroup.append([[k],[sr/len(i)]])

        for i in range(len(everygroup)):
            for j in range(0,len(everygroup)-i-1):
                if everygroup[j][1] < everygroup[j+1][1]:
                    everygroup[j],everygroup[j+1] = everygroup[j+1], everygroup[j]
                    
        print('Расположение групп в поряке убывания по среднему баллу:')
        for i in range(len(everygroup)):
            print('Группа',*everygroup[i][0])


gr1 = [[5, 4, 3, 2, 1], [5, 4, 2, 3, 5],[5, 2, 1, 5, 3]]
gr2 = [[5, 4, 4, 4, 4], [5, 4, 4, 4, 5]]
gr3 = [[2, 5, 3, 2, 2], [2, 4, 5, 5, 5]]

table = grades(gr1, gr2, gr3)
table.adding()
```

```python
class teams(): 
    def __init__(self,team1,team2,team3): 
        self.team1 = team1 
        self.team2 = team2 
        self.team3 = team3 
        
    def result(self): 
        res = [] 
        results = [self.team1, self.team2, self.team3] 
        for i in results: 
            sums = 0 
            for j in i: 
                if j <= 5: 
                    sums += 6-j 
            res.append(sums) 
            
        mx = 0 
        for i in res: 
            mx = max(mx,i) 
            
        win = [] 
        for i in range(len(res)): 
            if res[i] == mx: 
                win.append(i+1) 
        if len(win) != 1: 
            for i in win: 
                if i == 1: 
                    for j in range(len(results[0])): 
                        if results[0][j] == 1: 
                            print('Победа команды',i) 
                if i == 2: 
                    for k in range(len(results[1])): 
                        if results[1][k] == 1: 
                            print('Победа команды',i) 
                if i == 3: 
                    for b in range(len(results[2])): 
                        if results[2][b] == 1: 
                            print('Победа команды',i) 
        else: 
            print('Победа команды',win[0]) 
                         
                 
gr1 = [int(i) for i in input('Введите занятые места, 6 штук через пробел:').split(' ')] 
gr2 = [int(i) for i in input('Введите занятые места, 6 штук через пробел:').split(' ')] 
gr3 = [int(i) for i in input('Введите занятые места, 6 штук через пробел:').split(' ')] 
borba = teams(gr1,gr2,gr3) 
borba.result() 
```
