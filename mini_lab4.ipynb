```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r'dataset.csv',sep=';')
```

```python
df.head()
```

```python
print(df.dtypes)
```

```python
df['price']=df['price'].str.replace(',','.')
df['price']=df['price'].astype(float)
```

```python
plt.scatter(df.area,df.price,color='red')
plt.xlabel('площадь(кв.м.)')                
plt.ylabel('стоимость(млн.руб)') 
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

СКОЛЬКО БУДЕТ СТОИТЬ КВАРТИРА 38 М^2?

```python
print(reg.predict([[38]]))
```

СКОЛЬКО БУДЕТ СТОИТЬ КВАРТИРА 200 М^2?

```python
print(reg.predict([[200]]))
```

```python
print(reg.predict(df[['area']]))
```

```python
print(reg.coef_)
print(reg.intercept_)
```

ПОСТРОИМ МОДЕЛЬ ЛИНЕЙНОЙ РЕГРЕССИИ

```python
plt.scatter(df.area,df.price,color='red')                                                                
plt.xlabel('площадь(кв.м.)')                  
plt.ylabel('стоимость(млн.руб)') 
plt.plot(df.area, reg.predict(df[['area']])) 
plt.show()
```

РАБОТА С ДРУГИМ ФАЙЛОМ

```python
pred = pd.read_csv(r'prediction.csv',sep=';')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
p = reg.predict(pred)
pred['predicted prices'] = p
print(pred)
```
